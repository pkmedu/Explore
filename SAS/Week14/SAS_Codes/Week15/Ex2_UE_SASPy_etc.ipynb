```python
#https://support.sas.com/software/products/university-edition/faq/jn_saspy_getstarted.htm
import saspy
sas=saspy.SASsession()
df= sas.sasdata2dataframe('cars', libref='sashelp')
df.describe()
```

```python
import pandas as pd
df_x = pd.DataFrame(df,columns=['Make','Type','Invoice'])
print(df_x)
```

```python
import numpy  as np
stats=df.groupby(['Make'])['Invoice'].aggregate(['count', 'min', 'max', 'mean', np.median])
print(stats)
```
