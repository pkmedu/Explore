##### index_col -> Makes passed column as index instead of 0, 1, 2, 3 ...

```python
import pandas as pd
df=pd.read_csv('c:/Users/pmuhuri/sascourse/Week14/DEMOGRAPHICS.dat',  index_col='NAME', sep=',')
df.head()
```

##### To select only the float columns,  you can use the select_dtypes method. 

```python
df.select_dtypes(include = ['float']).head()
```

#### To select the sixth row by index position in the df DataFrame, you can pass number 5 to the .iloc indexer.

```python
df.iloc[5]
```

#### To do the same thing, you can use the index label with the .loc indexer.

```python
df.loc['DOMINICAN REPUBLIC']
```

##### You can pass a list of NAME values (i.e., index labels) to the .loc indexer to create a subset of the DataFrame.

```python
rows=["CHINA", "INDIA"]
df.loc[rows]
```

#### Selecting disjointed rows and columns by index position 

To select particular columns and rows from DataFrame by index position specified in range, you can use the .iloc indexer.


```python
df.iloc[2:5,[5,8,9]]
```

##### Select multiple rows & columns by Labels in DataFrame using loc[]

```python
df.loc[['CANADA','COSTA RICA', 'CUBA'],['pop', 'totalFR', 'AdolescentFPpct']]
```
