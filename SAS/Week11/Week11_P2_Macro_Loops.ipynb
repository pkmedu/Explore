##### %DO-%END Construct is used to geneate SAS code
* inside of a macro
* I is the macro variable becaise it is in the %DO-%END loop.


```sas
options nodate nonumber nosource nonotes nosymbolgen;
ods html close;
%macro runit;
%local i;
%Do i = 1 %to  5;
   %put i = Test&i;
%end;
%mend runit;
%runit
```

##### %DO-%While Construct

```sas
options nodate nonumber nosource nonotes nosymbolgen;
ods html close;
ods html close;
%macro runit;
%local i;
%Do %while (&i lt 6);
      %let i = %eval(&i + 1);
       %put i = Test&i;
%end;
%mend runit;
%runit
```

##### %DO-%Until Construct

```sas
options nodate nonumber nosource nonotes nosymbolgen;
ods html close;
ods html close;
%macro runit;
%local i;
%Do %until (&i = 6);
      %let i = %eval(&i + 1);
       %put i = Test&i;
%end;
%mend runit;
%runit
```
